```python
from google.adk.agents import LlmAgent, ParallelAgent, SequentialAgent
from google.adk.tools import google_search
```

```python
GEMINI_MODEL = "gemini-2.0-flash"
```

```python
# Researcher 1: Renewable Energy
researcher_agent_1 = LlmAgent(
                            name = "RenewableEnergyResearcher",
                            model = GEMINI_MODEL,
                            instruction = """
                                            You are an AI Research assistant specializing in energy. Research the latest advancements in 
                                            'renewable energy sources'/ Use the Google search tool provided. Summarize your key findings concisely (1-2 sentences).
                                            Output *only* the summary.
                            """,
                            description = "Researches renewable energy sources.",
                            tools = [google_search],
                            output_key="renewable_energy_result"
                            )
```

```python
researcher_agent_2 = LlmAgent(
                            name = "EVReseacher",
                            model = GEMINI_MODEL,
                            instruction = """
                                            You are an AI Research assistant specializing in transportation. Research the latest developments in 
                                            'electrical vehicle technology'. Use the Google search tool provided. Summarize your key findings concisely (1-2 sentences).
                                            Output *only* the summary.
                            """,
                            description = "Researches electric vehicle technology.",
                            tools = [google_search],
                            output_key="ev_technology_result"
                            )
```

```python
researcher_agent_3 = LlmAgent(
                            name = "CarbonCaptureReseacher",
                            model = GEMINI_MODEL,
                            instruction = """
                                            You are an AI Research assistant specializing in climate solutions. Research the current status of 
                                            'carbon capture methods'. Use the Google search tool provided. Summarize your key findings concisely (1-2 sentences).
                                            Output *only* the summary.
                            """,
                            description = "Researches carbon captrure methods.",
                            tools = [google_search],
                            output_key="carbon_capture_result"
                            )
```

```python
parallel_research_agent = ParallelAgent(
                                        name = "ParallelWebResearchAgent",
                                        sub_agents = [
                                                        researcher_agent_1,
                                                        researcher_agent_2,
                                                        researcher_agent_3
                                                            
                                                      ],
                                        description = "Runs multiple research agents in parallel to gather information"
            
                                        )
```

```python
merger_agent = LlmAgent(
                          name = "SynthesisAgent",
                          model = GEMINI_MODEL,
                          instruction = """
                                              You are an AI assistant responsible for combining research findings into a structured report.
                                              Your primary task is to synthesize the following research summaries, clearly attributing findings
                                              to their source areas. Structure your response using headings for each topic. Ensure that the report is 
                                              coherent and integrated the key points smoothly.
                                              ** Cruciallu: Your entire reponse must be gorunded *exclusiovely* on the information provided in the
                                              'Input Summaries' below. Do NOT add any external knowledge, facts, or details not present in these
                                              specific summaries.**
                                              **Input Summaries:**
                                              *     **Renewable Energy:**
                                                    {renewable_energy_result}
                                              *     **Electric Vehicles:**
                                                    {ev_technology_result}
                                              *     **Carbon Capture:**
                                                    {carbon_capture_result}

                                             **Output Format:**
                                             ## Summary of Recent Sustainable Technology Advancements

                                             ### Renewable Energy Findings
                                             (Based on RenewableEneryResearcher's findings)
                                             [Synthesize and elaborate *only* on the renewable energy input summary provided above.]
                                             
                                             ### Electric Vehicle Findings (Based on EVResearcher's findings) 
                                             [Synthesize and elaborate *only* on the EV input summary provided above.] 
                                             
                                             ### Carbon Capture Findings (Based on CarbonCaptureResearcher's conclusions) 
                                             [Synthesize and elaborate *only* on the carbon capture input summary provided above.]  
                                              
                                              ### Overall Conclusion 
                                              [Provide a brief (1-2 sentence) concluding statement that connects *only* the findings presented above.] 
                                              Output *only* the structured report following this format. 
                                              Do not include introductory or concluding phrases outside this structure, 
                                              and strictly adhere to using only the provided input summary content. 
                                        """, 
                                        description="Combines research findings from parallel agents into a structured, cited report, strictly grounded on provided inputs."
                                             
                                            
                        )
```

```python
sequential_pipeline_agent = SequentialAgent(
                                               name = "ResearchAndSynthesisPipeline",
                                               sub_agents = [parallel_research_agent, merger_agent],
                                               description="Coordinates parallel research and synthesizes the result.s"              
    
                                            )
```

```python
root_agent = sequential_pipeline_agent
```

```python
import nest_asyncio
from google.adk.runners import InMemoryRunner
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService
from google.genai import types 
import uuid

async def main():
    query = "Summarize the latest developments in vector quantization for embeddings."

    Content = types.Content(
        role="user",
        parts=[types.Part(text=query)]
    )

    
    session_service = InMemorySessionService()
    app_name = "research_app"
    user_id = "user_1"
    session_id = f"sess_{uuid.uuid4().hex[:8]}"

    # Create the session once (important)
    await session_service.create_session(
                                        app_name=app_name,
                                        user_id=user_id,
                                        session_id=session_id,
                                        )

    # Many ADK setups use a Runner/AgentRunner class
    runner = Runner(agent=root_agent, 
                    app_name=app_name,
                    session_service=session_service,)

    #result = await runner.run(user_query)
    query = "Summarize the latest developments in vector quantization for embeddings."

    content = types.Content(
        role="user",
        parts=[types.Part(text=query)]
    )
    #  DO NOT await runner.run
    events = runner.run(
        user_id=user_id,
        session_id=session_id,
        new_message=content,
    )
    
    final_text = None
    for event in events:
        # ADK versions vary slightly here
        if hasattr(event, "is_final_response") and event.is_final_response():
            final_text = event.content.parts[0].text
        elif getattr(event, "type", None) == "FINAL_RESPONSE":
            final_text = event.content.parts[0].text

    print("\n=== FINAL RESULT ===\n")
    print(final_text)


    
    # Result is often a dict-like object or a message object
    print("=== FINAL RESULT ===")
    print(final_text)
```

```python


# Cell 11: Apply patch
nest_asyncio.apply()

# Cell 12: Run
await main()
```
